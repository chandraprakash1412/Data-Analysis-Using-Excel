```python
import pandas as pd                                                            # Importing pandas library   # "as" is used as alias
d={"col1":[1,2,3,4,5],"col2":[6,7,8,9,10],"col3": [11,12,13,14,15]}             # Creating dctionary 
df=pd.DataFrame(d)                                                            # creating pandas datafeame
print(df)                                                                     # printing dataframe
```

# counting row in python DataFrame
'''  SYNTAX: 
    name_of_dataframe.shape[0]                                  # [0] is used for rows   
    => this count the number of rows in dataframe '''

```python
e=df.shape[0]         
print("number of rows : ",e)
```

# counting columns in python DataFrame
'''  SYNTAX: 
    name_of_dataframe.shape[1]                                # [1] is used for columns
    => this count the number of columns in dataframe '''

```python
f=df.shape[1]         
print("number of columns : ",f)
```

```python
# Locate Row
''' => Pandas use the loc attribute to return one or more specified row(s)'''
```

```python
print(df.loc[0]) #print values of row 0
print()  # goes to newline
```

```python
print(df.loc[[0,1]]) #print values of row 0 & 1
```

# Named Indexes
''' => With the index argument, you can name your own indexes. '''

```python
f=pd.DataFrame(d,index=["a","b","c","d","e"])
print(f)
```

# Locate Named Indexes
'''Use the named index in the loc attribute to return the specified row(s).'''

```python
g=f.loc["a"]
print(g)
h=f.loc[["a","b"]]
print(h)
```

```python
# Load Files Into a DataFrame
''' => If your data sets are stored in a file, Pandas can load them into a DataFrame.'''
print("\n Reading CSV file :\n")
df2=pd.read_csv("d.csv")
print(df2)     # This statement will not print the entire dataframe if DataFrame is too large.
```

```python
# Tip: use to_string() to print the entire DataFrame.

print("\n Reading CSV file completely :\n")
print(df2.to_string())   # This will print the entire dataframe if dataframe is too large as well.
```

```python
# reading json file

print("\n Reading JSON file :\n")
data2=pd.read_json('data1.json')
print(data2)
#print(data2)

print("\n Reading complete JSON file :\n")
print(data2.to_string())
```

```python
# df.head()
''' The head() method returns the headers and a specified number of rows, starting from the top.
 => # header :- column names of DataFrame are known as headers. '''

# Note: if the number of rows is not specified, the head() method will return the top 5 rows.
print("Reading CSV file first 10 rows :\n")
print(df2.head(10))    # "df2.head(10)" : This will print starting 10 rows of dataframe with header

print(' \n start 5 rows of a file \n')
print(df2.head())        # "df2.head()" : This will print starting 5 rows of dataframe with header

# df.tail()

''' => There is a tail() method for viewing the last rows of the DataFrame.
    => The tail() method returns the headers and a specified number of rows, starting from the bottom. '''

print("\n printing last 5 rows of a file : \n")
print(df2.tail())     # "df2.tail()" : This will print last 5 rows of dataframe with header.  

print("\n printing last 10 rows of a file : \n")
print(df2.tail(10))     # "df2.tail(10)" : This will print last 10 rows of dataframe with header. 

#Info About the Data
'''
 => The DataFrames object has a method called info(), that gives you more information about the data set.'''

print("\n Info of data2 json file : \n")
print(data2.info())              # data2.info() will tell the information of data2 file.
```
